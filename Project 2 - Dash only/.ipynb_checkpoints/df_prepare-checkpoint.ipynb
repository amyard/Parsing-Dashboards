```python
import pandas as pd

import zipfile

fantasy_zip = zipfile.ZipFile('gtd.zip')
fantasy_zip.extractall()

fantasy_zip.close()
```

```python
df = pd.read_csv('globalterrorismdb_0718dist.csv', encoding='ISO-8859-1')
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','region_txt':'Region',
                       'attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed','nwound':'Wounded',
                       'summary':'Summary','gname':'Group','targtype1_txt':'Target_type','weaptype1_txt':'Weapon_type',
                       'motive':'Motive'},inplace=True)
df=df[['Year','Month','Day','Country','Region','city','latitude','longitude','AttackType','Killed','Wounded','Group','Target_type','Weapon_type']]
df['Weapon_type'] = df['Weapon_type'].apply(lambda x: x.replace(' (not to include vehicle-borne explosives, i.e., car or truck bombs)', ''))


df.head()
```

```python
df['inj'] = (df['Killed']+df['Wounded'])
abs = pd.DataFrame(df['inj'].value_counts()).reset_index().sort_values('index')
```

```python
def get_bla(x):
    if x==0:  return str('0')
    elif (x>0)&(x<6): return str('1-5')
    elif (x>=6)&(x<11):  return str('6-10')
    elif (x>=11)&(x<26): return str('11-25')
    elif (x>=26)&(x<51): return str('26-50')
    elif (x>=51)&(x<101): return str('51-100')
    elif (x>=101)&(x<251): return str('101-250')
    elif (x>=251)&(x<501): return str('251-500')
    else: '501-...'

df['inj-group'] = df['inj'].apply(lambda x: get_bla(x))
#df['inj'].dropna(axis = 0, inplace = True)
df.head()
```

```python
df = df[df['Month'] != 0]

df.dropna(subset = ['inj','latitude','longitude'], inplace = True)
df.sort_values('inj-group', inplace = True)
df.reset_index(drop=True, inplace = True)
df.head()
```

```python
text_full=[]
for i in range(len(df)):
    text = 'Date: {}/{}/{}<br>{}/{}<br>Killed: {}<br>Wounded:{}'.format(df.loc[i,'Day'],df.loc[i,'Month'],df.loc[i,'Year'],
                                                   df.loc[i,'Country'],df.loc[i,'city'],
                                                   df.loc[i,'Killed'].astype(int).round(),
                                                   df.loc[i,'Wounded'].astype(int).round())
    text_full.append(text)
df['text'] = text_full
df.head()
```

```python
df['Month'].replace({1:'Январь',2:'Февраль',3:'Март',4:'Апрель',5:'Май',6:'Июнь',7:'Июль',8:'Август',
              9:'Сентябрь',10:'Октябрь',11:'Ноябрь',12:'Декабрь'}, inplace = True)
df['Month'].unique()
```

```python
df['inj-group'].unique()
```

```python
df['Region'].unique()
df['New_region'] = df['Region'].replace({'Western Europe':'europe', 'Southeast Asia':'asia', 'South Asia':'asia', 'South America':'south america',
                                        'Sub-Saharan Africa':'africa','Eastern Europe':'europe', 'Middle East & North Africa':'africa',
                                        'Central Asia':'asia', 'East Asia':'asia','North America':'north america', 'Australasia & Oceania':'asia',
                                        'Central America & Caribbean':'north america'})
df['New_region'].unique()
```

```python
df.to_csv('for_dash.csv', index = False)
```
